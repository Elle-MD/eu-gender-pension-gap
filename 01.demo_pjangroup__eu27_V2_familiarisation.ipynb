## Missing values
Two distinct patterns identified:
1. Greece (EL): all age categories missing for 2011 only
2. EU27_2020 Y_GE85: missing 2013-2020 - aggregate not calculated by Eurostat until 2021. Individual country data available throughout.

```python
# Check missing values per column
df.isnull().sum() 
```

There are no missing values. 

```python
# Check if there are any duplicate rows
df.duplicated().sum()
```

There are no duplicate rows.

```python
# Preliminary flag scan across year columns
year_cols = [str(y) for y in range(2011, 2026)]
flagged = df[year_cols].apply(lambda x: x.str.contains('[a-zA-Z]', na=False)).sum().sum()
total = df[year_cols].size
print(f"Flagged values: {flagged} / {total} ({flagged/total*100:.1f}%)")
print(f"Clean values: {total-flagged} / {total} ({(total-flagged)/total*100:.1f}%)")
```

```python
# Check for Eurostat flags in values
for col in df.columns[1:]:
    flagged = df[col].str.contains('[a-zA-Z]', na=False)
    if flagged.any():
        print(f"{col}: {flagged.sum()} flagged values")
```

```python
# See what types of flags appear per year
for col in year_cols:
    flags = df[col].str.extract(r'([a-zA-Z]+)')[0].dropna()
    print(f"{col}: {flags.value_counts().to_dict()}")
```

```python
# Confirm it's the same 24 rows across all flag types
mask = df['2023'].str.contains('[a-zA-Z]', na=False)
print(df[mask][df.columns[0]].value_counts())
```

## Summary

**Dataset:** demo_pjangroup — Population by age group and sex
- **Shape:** 672 rows x 16 columns
- **Period:** 2011-2025
- **Geographies:** 27 EU countries + EU27_2020 aggregate

**Key findings:**
- Combined first column needs splitting into 5 separate columns
- All year columns read as object type due to flags mixed into values
- Leading/trailing whitespace in both column names and values
- Missing values coded as ":" not NaN — pandas doesn't detect automatically
- Greece (EL) missing all 2011 data
- EU27_2020 Y_GE85 missing 2013-2020 — aggregate not available until 2021
- Some values carry Eurostat flags (p=provisional, e=estimated, ep=estimated provisional. Figures are retained as official Eurostat estimates and treated as comparable to final values.

**Actions required in cleaning:**
- Strip whitespace from column names and values
- Split combined first column
- Replace ":" with NaN
- Separate flags from values
- Melt to long format
